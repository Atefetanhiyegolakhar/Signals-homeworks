```python
import numpy as np
import matplotlib.pyplot as plt

def rect(t):
    return np.where(np.abs(t) <= 0.5, 1, 0)

t = np.linspace(-2, 2, 1000)
x = rect(t)

plt.figure(figsize=(10, 6))
plt.plot(t, x, 'b-', linewidth=2)
plt.title('Rectangular Pulse Π(t)')
plt.xlabel('t')
plt.ylabel('Π(t)')
plt.grid(True)
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rect(t):
    return np.where(np.abs(t) <= 0.5, 1, 0)

t = np.linspace(-5, 5, 1000)

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
x1 = rect(3*t - 0.5)
plt.plot(t, x1, 'b-', linewidth=2)
plt.title('x(3t - 0.5)')
plt.grid(True)
plt.ylim(-0.5, 1.5)

plt.subplot(3, 1, 2)
x2 = rect(0.5*t + 3)
plt.plot(t, x2, 'r-', linewidth=2)
plt.title('x(0.5t + 3)')
plt.grid(True)
plt.ylim(-0.5, 1.5)

plt.subplot(3, 1, 3)
x3 = rect(t - 2) - t
plt.plot(t, x3, 'g-', linewidth=2)
plt.title('x(t - 2) - t')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-2, 3)
x = np.array([0, 1, 1, 1, 0])

plt.figure(figsize=(10, 6))
plt.stem(n, x, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_original = np.arange(-2, 3)
x_original = np.array([0, 1, 1, 1, 0])

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
n1 = n_original + 2
x1 = x_original
plt.stem(n1, x1, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('x[n-2]')
plt.grid(True)
plt.ylim(-0.5, 1.5)

plt.subplot(3, 1, 2)
n2 = np.arange(-3, 0)
x2 = np.array([1, 1, 0])
plt.stem(n2, x2, linefmt='r-', markerfmt='ro', basefmt='r-')
plt.title('x[1.5n+3]')
plt.grid(True)
plt.ylim(-0.5, 1.5)

plt.subplot(3, 1, 3)
n3 = np.arange(-2, 2)
x3 = np.array([-3, -2, -2, -2])
plt.stem(n3, x3, linefmt='g-', markerfmt='go', basefmt='r-')
plt.title('x[n] + x[2n] - 3')
plt.grid(True)
plt.ylim(-3.5, -1.5)

plt.tight_layout()
plt.show()
```
