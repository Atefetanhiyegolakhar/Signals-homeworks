```python

import numpy as np
import matplotlib.pyplot as plt


B = 1.0
w = 2 * np.pi
Phi = 0.0


t = np.linspace(0, 5, 2000)


a_neg = -0.7
a_pos = 0.7


x_neg = B * np.exp(a_neg * t) * np.cos(w * t + Phi)
x_pos = B * np.exp(a_pos * t) * np.cos(w * t + Phi)


plt.figure(figsize=(8, 3.5))
plt.plot(t, x_neg, color='blue')
plt.title(" for (a < 0)")
plt.xlabel("t (s)")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 3.5))
plt.plot(t, x_pos, color='red')
plt.title(" for (a > 0)")
plt.xlabel("t (s)")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()
```
